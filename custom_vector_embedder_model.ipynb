```python
import gensim
import os
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
all_text = []
file_loc = open("/kaggle/input/game-of-thrones-books/003ssb.txt")
corpus = file_loc.read()
sentence_split_corpus = sent_tokenize(corpus)
for sentence in sentence_split_corpus:
    all_text.append(simple_preprocess(sentence))     
```

```python
all_text[0:5]
```

```python
len(all_text)
```

## embedder model

```python
model = gensim.models.Word2Vec(
    window = 10,
    min_count = 2,
)
```

```python
model.build_vocab(all_text)
```

```python
model.train(all_text,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.wv.most_similar('daenerys')
```
