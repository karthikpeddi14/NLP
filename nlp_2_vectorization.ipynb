### 2. Bag of words

```python
import pandas as pd
df = pd.DataFrame({'text' : ['one two three', "two three one", "one one one"], 'output':[1,1,0]})
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df.text)
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

```python
cv.transform(["two two four"]).toarray()
```

### 4. TFIDF vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
bow_tfidf = tfidf.fit_transform(df.text)
```
